# Linear Models with Different Basis for a 1D input Space

In this exercise you will see develop an intuition of how the different weights affect the basis via visualisation. You will get familiar with Gaussian Basis (aka Radial Basis functions) as well as the sigmoid and polynomials.
We will start from 1d and move into 2d. As always visualisation with high dimension space > 3 is not possible so the exercise is meant to give you an idea of the behaviour of the basis function.

**Learning Outcome**
1. Understand the behaviour of polynomial basis
2. Understand the behaviour of radial basis function
3. Understand the behaviour of logistic sigmoid
4. Ability to evaluate the effect of changing the weights on the different basis functions


## Linear Model with a Linear Basis

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(-3,3,100)
w0 = 1
w1 = 2
y  = w0 + w1*x1

plt.plot(x1, y, '-r')
plt.grid()
```

## Two Linear Models with Linear Basis Same Slope

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3,3,100)

w0, w1 = 1,2
v0, v1 = 4,2

y1 = w0 + w1*x
y2 = v0 + v1*x

plt.plot(x, y1, '-r', label='y1='+str(w0)+'+'+str(w1)+'x')
plt.plot(x, y2, '-b', label='y2='+str(v0)+'+'+str(v1)+'x')

plt.legend(loc='upper right')
plt.grid()
```

## Linear Models with Linear Basis

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3,100)
w0, w1 = 0,2
v0, v1 = 5,2
b0, b1 = 5,4

y1 = w0 + w1*x
y2 = v0 + v1*x
y3 = b0 + b1*x

# set up the figure box to a traditional two axeses
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# now plot...
plt.plot(x, y1, '-r', label='y1='+str(w0)+'+'+str(w1)+'x')
plt.plot(x, y2, '-b', label='y2='+str(v0)+'+'+str(v1)+'x')
plt.plot(x, y3, '-g', label='y3='+str(b0)+'+'+str(b1)+'x')

plt.title('Linear Models $ y = w_0 + w_1x $')
plt.legend(loc='upper right')

plt.savefig('UnivariateLinearModel.png')
```

## Radial Basis Functions with Different Means and Variances

```python
import numpy as np
import matplotlib.pyplot as plt

# Samples first
x = np.linspace(-3, 5, num=100)

def RBF(x, mean, variance):
    # Radial Basis Function
    return (np.exp(-(x - mean)**2 / (2 * variance)))
mu1, var1 = 0,1
mu2, var2 = 0,0.4
mu3, var3 = 2,4

fig = plt.figure(figsize=(5, 3))
plt.plot(  x,  RBF(x, mean=mu1, variance=var1), '-r',  label='$e^{-\dfrac{(x-'+str(mu1) +')^2}{'+str(2*var1)+'}}$')
plt.plot(  x,  RBF(x, mean=mu2, variance=var2), '-b',  label='$e^{-\dfrac{(x-'+str(mu2) +')^2}{'+str(2*var2)+'}}$')
plt.plot(  x,  RBF(x, mean=mu3, variance=var3), '-g',  label='$e^{-\dfrac{(x-'+str(mu3) +')^2}{'+str(2*var3)+'}}$')

plt.xlabel('$x$', fontsize=13)
plt.ylabel('$\phi(x)$', fontsize=13)

plt.title('RBF Examples with different means and variances')

plt.ylim([0, 1])
plt.xlim([-3, 5])

plt.legend(loc=5)
fig.subplots_adjust(bottom=0.15)
plt.savefig('RBF.png')
```

## Now we tackle Sigmoidal Functions

```python
import numpy as np
import matplotlib.pyplot as plt

# Samples first
x = np.linspace(-15, 15, num=100)

def Sig(x, mean, variance):
    # Sigmoidal Basis Function
    return 1/(1+(np.exp(-(x - mean) / (variance))))
mu1, var1 = 0,1
mu2, var2 = 0,0.4
mu3, var3 = 2,4

# set up the figure box to a traditional two axeses
fig = plt.figure(figsize=(5, 3))
ax  = fig.add_subplot(1, 1, 1)
ax.spines['left'].  set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['right']. set_color('none')
ax.spines['top'].   set_color('none')

# Now plot
plt.plot(  x,  Sig(x, mean=mu1, variance=var1), '-r',  label='$\dfrac{1}{1+e^{-\dfrac{(x-'+str(mu1) +')}{'+str(2*var1)+'}}}$')
plt.plot(  x,  Sig(x, mean=mu2, variance=var2), '-b',  label='$\dfrac{1}{1+e^{-\dfrac{(x-'+str(mu2) +')}{'+str(2*var2)+'}}}$')
plt.plot(  x,  Sig(x, mean=mu3, variance=var3), '-g',  label='$\dfrac{1}{1+e^{-\dfrac{(x-'+str(mu3) +')}{'+str(2*var3)+'}}}$')

plt.xlabel('$x$', fontsize=13)
plt.ylabel('$\phi(x)$', fontsize=13)

plt.title('Sigmoidal Basis Functions')

plt.ylim([0, 1])
#plt.xlim([-3, 5])

plt.legend(loc=5)
fig.subplots_adjust(bottom=0.15)
plt.savefig('Sigmoidal.png')
```
